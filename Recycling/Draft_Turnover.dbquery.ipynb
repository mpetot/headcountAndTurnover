```sql
SELECT
  EPMBRANCH,
  e.description,
  sum(value)
FROM
  (
    SELECT
      to_date(CAST(DateInt as string), 'yyyyMMdd') as dateAsDate,
      CostType,
      EPMBranch,
      '' as MetricType,
      '' as jobcodecategory,
      0 as booOverScale,
      Round(FTEVoluntaryTurnover, 3) as value,
      'Turnover' as source
    from
      `na-cdp-bu-hr-dev`.curated.voluntaryturnover
    WHERE
      DateInt >= 20231001
      and FTEVoluntaryTurnover <> 0
  ) LEFT JOIN (SELECT
    EntityId,
    Description
  FROM
    `na-cdp-bu-finance-dev`.curated.v_entity) as e on e.entityid = epmbranch
GROUP BY
  ALL
```
