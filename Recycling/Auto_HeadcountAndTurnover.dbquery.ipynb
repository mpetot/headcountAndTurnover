```sql
With fte as (
  SELECT
    to_date(CAST(DateInt as string), 'yyyyMMdd') as dateAsDate,
    CostType,
    EPMBranch,
    BreakDownTwo as MetricType,
    JobCodeCategory,
    CASE
      WHEN isoverscale = 'Y' then 1
      ELSE 0
    END booOverScale,
    Round(FTE, 3) as value,
    'FTE' as source
  from
    `na-cdp-bu-hr-dev`.curated.fte
  WHERE
    DateInt >= 20231001
    and FTE <> 0
    and FTE is not null
),
usdaamt as (
  SELECT
    to_date(CAST(DateInt as string), 'yyyyMMdd') as dateAsDate,
    CostType,
    EPMBranch,
    BreakDownTwo as MetricType,
    CostType as jobcodecategory,
    0 as booOverScale,
    Round(AmountAnnual, 3) as value,
    'USDAmount' as source
  FROM
    `na-cdp-bu-hr-dev`.curated.usdamount
  WHERE
    DateInt >= 20231001
    and AmountAnnual <> 0
),
turnover as (
  SELECT
    to_date(CAST(DateInt as string), 'yyyyMMdd') as dateAsDate,
    CostType,
    EPMBranch,
    '' as MetricType,
    '' as jobcodecategory,
    0 as booOverScale,
    Round(FTEVoluntaryTurnover, 3) as value,
    'Turnover' as source
  from
    `na-cdp-bu-hr-dev`.curated.voluntaryturnover
  WHERE
    DateInt >= 20231001
    and FTEVoluntaryTurnover <> 0
)
SELECT
  *
FROM
  fte
UNION ALL
SELECT
  *
FROM
  usdaamt
UNION all
SELECT
  *
from
  turnover;






SELECT
  DateInt,
  EPMBranch,
  IsOverScale,
  count(*) cntrows
from
  `na-cdp-bu-hr-dev`.curated.fte
GROUP BY
  ALL

  ORDER BY 1 ASC, 2 ASC;



```
